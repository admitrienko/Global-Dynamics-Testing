```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

datadir = '/Users/sbittner/Documents/wormies/data/';
fname = datadir + 'kalman_times.npz';

npzfile = np.load(fname);

em_times = npzfile['em_times'];
ll_times = npzfile['ll_times'];
n_time_pts_list = npzfile['n_time_pts_list'];
n_dim_state_list = npzfile['n_dim_state_list'];

n_time_pts_len = len(n_time_pts_list)
n_dim_state_len = len(n_dim_state_list)




```

```python
time_pts = np.array(n_time_pts_list);

process_str_list = ['EM', 'log-lik'];
times_list = [em_times, ll_times];
n_processes = len(process_str_list);

fontsize = 14;
figsize = (4*n_processes, 4);
fig = plt.figure(figsize=figsize);

for plot_ind in range(n_processes):
    process_str = process_str_list[plot_ind];
    times = times_list[plot_ind];
    
    ax = fig.add_subplot(1,n_processes,plot_ind+1);
    labels = [];
    for j in range(n_dim_state_len):
        n_dim_state = n_dim_state_list[j];
        label = '%d latent dims' % n_dim_state;
        plt.plot(time_pts, times[:,j], label = label);
    plt.xlabel('# time points', fontsize=14);
    plt.ylabel('seconds per iteration', fontsize=14);
    plt.title('%s complexity' % process_str, fontsize = 16);
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.legend();
plt.show();
```

```python
dim_state = np.array(n_dim_state_list);

fig = plt.figure(figsize=figsize);


for plot_ind in range(n_processes):
    process_str = process_str_list[plot_ind];
    times = times_list[plot_ind];
    
    ax = fig.add_subplot(1,n_processes,plot_ind+1);
    
    labels = [];
    for i in range(n_time_pts_len):
        n_time_pts = n_time_pts_list[i];
        label = '%d time points' % n_time_pts;
        plt.plot(dim_state, times[i,:], label=label);

    plt.xlabel('# time points');
    plt.ylabel('seconds per iteration');
    plt.title('%s complexity' % process_str, fontsize = 16);
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.legend();
plt.show();
```
